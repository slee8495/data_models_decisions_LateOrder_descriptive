```python
import pandas as pd
from faker import Faker


# Load spreadsheet
xl = pd.ExcelFile('anonymized_data.xlsx')

# Load a sheet into a DataFrame by its name
df = xl.parse('Sheet1')
df

fake = Faker()
```

```python
def anonymize_df(df):
    columns_to_anonymize = ['profile_owner', 'leader_name', 'loc', 'order', 'customer']

    for column in columns_to_anonymize:
        unique_values = df[column].unique()
        if column == 'loc':
            fake_values = [fake.random_int(min=1, max=50) for _ in range(len(unique_values))]
        elif column == 'order':
            fake_values = [fake.random_int(min=1000, max=50000) for _ in range(len(unique_values))]
        elif column == 'customer':
            fake_values = [fake.company() for _ in range(len(unique_values))]
        elif df[column].dtype == 'object':
            fake_values = [fake.unique.name() for _ in range(len(unique_values))]
        else:
            fake_values = [fake.unique.random_number() for _ in range(len(unique_values))]
        mapping_dict = dict(zip(unique_values, fake_values))
        df[column] = df[column].map(mapping_dict)

    return df

fake.unique.clear()
df = anonymize_df(df)
```
